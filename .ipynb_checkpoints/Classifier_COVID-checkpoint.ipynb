## Importing Required Packages

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, roc_auc_score,cohen_kappa_score
from skimage.io import imread, imshow
from skimage import filters
from skimage.filters import prewitt_h,prewitt_v
from skimage.feature import hog
from skimage.transform import rescale
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, cross_validate
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import BaggingClassifier
import seaborn as sns
```

## Setting up data extractor

```python
def get_files(path):
    file_list = []
    for path, directories, files in os.walk(path):
        file_list = files
        return file_list
```

```python
img_raw = cv2.imread(f'COVID19_Lung_CT/CT_COVID/2019-novel-Coronavirus-severe-adult-respiratory-dist_2020_International-Jour-p3-89%0.png',0)
print(img_raw.shape)
a = np.array(img_raw).flatten()
print(a.shape)
```

## Extracting COVID-19 data, normalizing & then adding to dataframe

```python
c_path = "./COVID19_Lung_CT/CT_COVID"
c_files = get_files(c_path)
c_data = []


for img in c_files:
    img_raw = imread(f'{c_path}/{img}',as_gray=True)
    try:
        img_norm = cv2.resize(img_raw,(100,100)) # This needs to be improved with effective feature extraction
        img_flat = np.array(img_norm)
        c_data.append([img_flat,1]) # Assigning label 1 for COVID data
    except AssertionError: # Using try except clause makes sure we don't have any images which would not resize properly
        pass
        
print(len(c_data))
```

```python
#pd.set_option("max_rows", None)
df = pd.DataFrame(c_data, columns = ['img_data', 'label'])
df['img_data'][0].shape
```

## Extracting Non COVID-19 data, normalizing & then adding to dataframe

```python
nc_path = "./COVID19_Lung_CT/CT_NonCOVID"
nc_files = get_files(nc_path)
nc_data = []

for img in nc_files:
    img_raw = imread(f'{nc_path}/{img}', as_gray=True)
    try:
        img_norm = cv2.resize(img_raw,(100,100))
        img_flat = np.array(img_norm)
        nc_data.append([img_flat,0])
    except AssertionError:
        pass
    
print(len(nc_data))
```

```python
df2 = pd.DataFrame(nc_data, columns = ['img_data', 'label'])
df2
```

```python
plt.imshow(df2['img_data'][0], cmap='gray')
```

### Merging data frames

```python
img_master = df.append(df2, ignore_index=True)
img_master
```

```python
img_final = img_master.sample(frac=1).reset_index(drop=True) #Shuffling dataframe and resetting index values
img_final
```

```python
plt.imshow(img_final['img_data'][0], cmap='gray')
```

```python
len(img_final)
```

### Applying feature extraction layer with HOG

```python
count = 0
final_data = []
for index, row in img_final.iterrows():
    hog_features, hog_example = hog(row['img_data'], pixels_per_cell=(10,10), block_norm = 'L2-Hys', visualize = True )
    final_data.append([hog_example, row['label']])

final_df = pd.DataFrame(data = final_data, columns = ['img_data', 'label'])
final_df
```

## Training, Testing & Validation Split

```python
import numpy as np
import pandas as pd

def train_validate_test_split(df, train_percent=.7, validate_percent=.1, seed=None):
    perm = np.random.permutation(df.index)
    m = len(df.index)
    train_end = int(train_percent * m)
    validate_end = int(validate_percent * m) + train_end
    train = df.iloc[perm[:train_end]]
    validate = df.iloc[perm[train_end:validate_end]]
    test = df.iloc[perm[validate_end:]]
    return train, validate, test


train, validate, test = train_validate_test_split(final_df)
```

### Training Data

```python
X_t = np.array(train['img_data'])
y_train = np.array(train['label']).flatten()


X_t1 = []

for i in X_t:
    X_t1.append(i)
    
X_t1 = np.array(X_t1)
X_train = np.reshape(X_t1, (522,100*100))

print(f'X_train.shape = {X_train.shape}')
print(f'y_train.shape = {y_train.shape}')
```

### Validation Data

```python
X_v = np.array(validate['img_data'])
y_validate = np.array(validate['label']).flatten()


X_v1 = []

for i in X_v:
    X_v1.append(i)
    
X_v1 = np.array(X_v1)

X_validate = np.reshape(X_v1, (74,100*100))

print(f'X_validate.shape = {X_validate.shape}')
print(f'y_validate.shape = {y_validate.shape}')
```

### Testing Data

```python
X_te = np.array(test['img_data'])
y_test = np.array(test['label']).flatten()


X_te1 = []

for i in X_te:
    X_te1.append(i)
    
X_te1 = np.array(X_te1)

X_test = np.reshape(X_te1, (150,100*100))

print(f'X_test.shape = {X_test.shape}')
print(f'y_test.shape = {y_test.shape}')
```

```python
print(f'X_train.shape = {X_train.shape}')
print(f'y_train.shape = {y_train.shape}')
print(f'X_validate.shape = {X_validate.shape}')
print(f'y_validate.shape = {y_validate.shape}')
print(f'X_test.shape = {X_test.shape}')
print(f'y_test.shape = {y_test.shape}')
```

## Validation of multiple models

### Validation Approach includes developing multiple models and then running various validation checks

### Implementing SVM Model

```python
sv = SVC(C=1,kernel='linear',gamma = 'auto')
```

```python
sv.fit(X_train, y_train)
```

```python
predict = sv.predict(X_validate)
```

#### Simple Accuracy Metric

```python
print(f'Model accuracy is {accuracy_score(y_validate, predict)*100:.2f}%')
```

#### Using K-Fold cross validation instead of hold-out approach to account for bias and deciding model

```python
kf = KFold(n_splits= 5, shuffle=True) #Defining number of splits for training dataset to set up validation and training split

sv = SVC(C=1,kernel='linear',gamma = 'auto') #Declaring model that will be used to train and validate

cv_results = cross_validate(sv, X_train, y_train, cv=kf, return_train_score=False) # Running K-Fold cross validation on Training dataset
```

```python
scoree = cv_results['test_score'].mean()*100
print(f'Model accuracy with K-Folds is {scoree:.2f}')
```

#### Using Confusion Matrix to account for more accuracy metrics such as precision, recall & F1-score

```python
print(f'{classification_report(y_validate, predict)}')
```

## Final Evaluation

### Looking at results SVM works best given the fact it is a high-margin classifier leading to better decision boundaries for our COVID-19 Binary classifier

```python
sv = SVC(C=1,kernel='linear',gamma = 'auto') #Redeclaring model since variables reused earlier
sv.fit(X_train, y_train)

bag_model = BaggingClassifier(base_estimator = sv, n_estimators = 10, bootstrap=True, random_state = 20)
bag_model.fit(X_train, y_train)
```

#### Accuracy Metric

```python
predict = bag_model.predict(X_test)
print(f'Model accuracy is {accuracy_score(y_test, predict)*100:.2f}%')
```

#### Classification Report (consists of precision (Type I Error), recall (Type II Error) and F1 Score)

```python
print(f'{classification_report(y_test, predict)}')
```

#### Confusion Matrix to calulcate TPR, FPR

```python
tn, fp, fn, tp = confusion_matrix(y_test, predict).ravel()
confusion = confusion_matrix(y_test, predict)

false_positive_rate = fp / (fp + tn) 
true_positive_rate = tp / (tp+fn)
false_negative_rate = fn / (tp + fn)


print(f"True Positive Rate is {true_positive_rate*100:.2f}%")
print(f"False Positive Rate is {false_positive_rate*100:.2f}%")
print(f"False Negative Rate is {false_negative_rate*100:.2f}%")

confusion_df = pd.DataFrame(confusion, index=range(2), columns=range(2))
axes = sns.heatmap(confusion_df, annot=True, fmt="d",cmap='YlGnBu')
```

### ROC Curve Area Under Curve

```python
roc_auc = roc_auc_score(y_test, predict)
print(f"Area under curve is {roc_auc*100:.2f}%")
```
